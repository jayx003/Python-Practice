```python
#3. Write a Function to Calculate Factorial

num =int(input('Enter num: '))
def factorial(num):
  if num == 0:
    return 1
  return num*factorial(num-1)

print('factorial of',num,':',factorial(num))
```

```python
#4. Write a Function to Check Prime Number

num =int(input('Enter num: '))
def prime(num):
  if num > 1:
    for i in range(2,num):
      if num%i == 0:
        print('Not Prime')
        break
    else:
      print('Prime')
prime(num)
```

```python
#6. Write a Function to Count Vowels in a String

name = input('Enter Name: ')
def vowels(name):
  count = 0
  for i in name:
    if i in 'aeiouAEIOU':
      count += 1
  print(count)
vowels(name)
```

```python
#7. Write a Function to Check Palindrome String
name = 'kayak'
def palindrome(name):
  if name[::-1] == name:
    print('Palindrome string')
  else:
    print('not a palindrome string')

palindrome(name)
```

```python
#8. Write a Function to Generate Fibonacci Series
#Problem Statement: Create a function that returns the Fibonacci series for n terms.
n = int(input('Enter n: '))

def factoriaL(n):
  series = []
  a,b = 0,1
  count = 0
  while count < n:
    series.append(a)
    a,b = b,a+b
    count += 1
  print(series)
factoriaL(n)
```

```python
#9. Write a Function to Remove Duplicates from a List
#Problem Statement: Create a function that removes duplicate values from a list while maintaining order.

l = [10,20,30,40,50,60,203,44,30,20,40,50,]
print(list(set(l)))

def convert(l):
  l = set(l)
  print(l)
convert(l)

# Order Maintained
def conversion(l):
  result = []
  for i in l:
    if i not in result:
      result.append(i)
  print(result)
conversion(l)
```

```python
#10. Write a Function to Sum Any Number of Arguments
#Problem Statement: Create a function that accepts any number of numeric arguments and returns their sum.

def total(*Numbers):
  total = 0
  for num in Numbers:
    total = total + num
  print(total)
total(10,20,30,40)
```
