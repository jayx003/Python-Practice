```python
#1. Write a Function to Check Even or Odd

num = int(input('Enter a number: '))
def evenodd(num):
  if num%2 == 0:
    print('Even')
  else:
    print('odd')

evenodd(num)
```

```python
#3. Write a Function to Calculate Factorial

num =int(input('Enter num: '))
def factorial(num):
  if num == 0:
    return 1
  return num*factorial(num-1)

print('factorial of',num,':',factorial(num))
```

```python
#4. Write a Function to Check Prime Number

num =int(input('Enter num: '))
def prime(num):
  if num > 1:
    for i in range(2,num):
      if num%i == 0:
        print('Not Prime')
        break
    else:
      print('Prime')
prime(num)
```

```python
#5. Write a Function to Reverse a String

name = input('Enter your nmae: ')

def rev(name):
  print(name[::-1])
rev(name)
```

```python
#6. Write a Function to Count Vowels in a String

name = input('Enter Name: ')
def vowels(name):
  count = 0
  for i in name:
    if i in 'aeiouAEIOU':
      count += 1
  print(count)
vowels(name)
```

```python
#7. Write a Function to Check Palindrome String
name = 'kayak'
def palindrome(name):
  if name[::-1] == name:
    print('Palindrome string')
  else:
    print('not a palindrome string')

palindrome(name)
```
