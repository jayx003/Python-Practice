### Function with arguments

```python
def operation(x,y):
  print('addition', x+y)
  print('subtraction', x-y)
  print('multiplication', x*y)
  print('division', x/y)
operation(int(input('x')), int(input('y')))
```

# **Types of arguments**
1. positional argument - (Sequence order matters)
2. Keyword argument - (Doesn't required to follow order, coz it directly refer key = valur format)
3. Default argumrnt- (When user is not supplying the values from calling then system will take these default values)
4. Variable- length argument - ()
  i. variable length positional args
  ii.


```python
#positional argument - (Sequence order matters)

def test(x,y,z):
  print(x,y,z)

test(10,20,30,)
test(z = 30,y = 20,x = 10)
```

```python
def info(name,age,sal):
  print('your name is',name,'your age is',age,'your salary',sal)

info('Jay',25,100000)
info(25,'jay',2500000)
```

```python
#2. Keyword argument - (Doesn't required to follow order, coz it directly refer key = value format)
info(age = 25,name ='jay',sal = 2500000)
```

```python
#3.Default argument
def info(name = 'Jay', age= 25, sal = 100000):
  print(name,age,sal)
info()
```

```python
#4. Variable- length argument - ()
 # i. variable length positional args
def test(**n):
  print(n)
test(x = 'jay')
```

```python
test(10,20,30)
```

```python
#ii. variable length keyword arguments
def test(**x):
  print(x)
test(x= 1, y='Jay',)
```

## Function with return keyword

```python
#function will work on specific requirement
#it will generate one fucntion
#thise result will be used as an input to the outside of the function
# for this purpose we need to use return keyboard

def test(balance):
  if balance <= 0:
    print('No')
  else:
    print('Yes')
result_test = test(int(input('balance: ')))
if result_test == 'yes':
  print('you can perform transaction')
else:
  print('you can not perform transaction')
```

```python
items = {'rice':60, 'wheat':50,'Milk': 40, 'Pen': 10}
bill = []

def purchase(*args):
  for item in args:
    print(item,items[item])
    bill.append(items[item])
  return bill
def check(total):
  return sum(total)
#calling of purchase
total = purchase('rice','wheat')
print(total)

ans = check(total)
print('Final bill amount is: ', ans)

```

```python
#calculate % and calculate grade
def percentage(marks_obtainded, total_marks):
  return (marks_obtainded/total_marks)*100

def grade(percentage):
  if percentage >= 90:
    return 'A'
  elif percentage >= 80:
    return 'B'
  elif percentage >= 70:
    return 'C'
  elif percentage >= 60:
    return 'D'
  else:
    return 'Fail'

per = percentage(int(input('Enter marks obtaines: ')),500)
grade(per)
print(per)
print(grade(per))
```
