```python
#1. Add Two Numbers Using Lambda
#Problem Statement: Write a lambda function to add two numbers.

a = 5
b = 6
add = (lambda a,b: a+b)
add(3,5)
```

```python
#2. Find Square and Cube Using Lambda
#Problem Statement: Create lambda functions to find the square and cube of a number.

sqr_cube = (lambda x: (x**2, x**3))
sqr_cube(3)

sqr = lambda x: x**2
cube = lambda x: x**3

print('Square of', x, 'is', sqr(4), 'and cube of',x,'is',cube(4))
```

```python
#3. Check Positive, Negative, or Zero
# Problem Statement: Write a lambda function to check whether a number is positive, negative, or zero.

check = lambda n: 'positive' if n > 0 else 'negative' if n<0 else 'Zero'
print(check(5), check(-5), check(0))
```

```python
#4. Filter Words Starting with a Specific Letter
#Problem Statement: Use lambda and filter() to select words starting with the letter A.
check = lambda word: 'pass' if word[0] == 'A' else 'fail'
print(check('Akash'))

words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
verify = list(filter(lambda word: word.startswith('A'), words))
verify
```

```python
#5. Convert List of Strings to Uppercase
#Problem Statement: Use lambda and map() to convert a list of strings into uppercase.

names = ["amit", "neha", "ravi"]
print(list(map(lambda x: x.upper(), names)))
```

```python
#6. Sort Students by Marks in Descending Order
#Problem Statement: Sort student records by marks from highest to lowest using lambda.
students = [("Amit", 85), ("Neha", 92), ("Ravi", 78)]
result = sorted(students, key=lambda x: x[1], reverse=True)
print(result)
```
